#Import the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

#loading the dataset

```python
data = pd.read_csv('/content/sample_data/Unemployment_Rate_upto_11_2020.csv')
data
```

#Checking dataset information

```python
data.info()
```

#Describing the dataset

```python
data.describe()
```

#check null/missing values

```python
data.isnull().sum()
```

#rename columns

```python
data.columns =['States','Date','Frequency','Estimated Unemployment Rate',
               'Estimated Employed','Estimated Labour Participation Rate',
               'Region','longitude','latitude']
```

#analysis top rows of dataset

```python
data.head()
```

#checking the correlaion between the feature of dataset

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(8, 6))

# Create the heatmap axis
ax = sns.heatmap(data.corr(), annot=True, linewidths=3)

# Adjust tick parameters on the heatmap
ax.tick_params(size=10, color='w', labelsize=10, labelcolor='w')

plt.show()
```

#estimated no of employees according to different region of india

```python
plt.title('Indian Unemployment')
sns.histplot(x='Estimated Employed', hue='Region', data=data)
plt.show()
```

#unemployment rate according to different region of india

```python
plt.figure(figsize=(10,8))
plt.title('Indian Unemployment')
sns.histplot(x='Estimated Employed', hue='Region', data=data)
plt.show()
```

#Dashboard to analyse the unemloyment rate of each indian state

```python
unemployment = data[['States', 'Region', 'Estimated Unemployment Rate']]
figure = px.sunburst(unemployment, path=['Region', 'States'],
                     values='Estimated Unemployment Rate',
                     width=700, height=600, color_continuous_scale='RdYlGn',
                     title="Unemployment Rate in India")
figure.show()
```
